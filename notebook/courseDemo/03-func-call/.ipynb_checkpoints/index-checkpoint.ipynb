````python
def get_completion(messages, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,  # æ¨¡å‹è¾“å‡ºçš„éšæœºæ€§ï¼Œ0 è¡¨ç¤ºéšæœºæ€§æœ€å°
        functions=[{  # ç”¨ JSON æè¿°å‡½æ•°ã€‚å¯ä»¥å®šä¹‰å¤šä¸ªï¼Œä½†æ˜¯åªæœ‰ä¸€ä¸ªä¼šè¢«è°ƒç”¨ï¼Œä¹Ÿå¯èƒ½éƒ½ä¸ä¼šè¢«è°ƒç”¨
            "name": "sum",
            "description": "è®¡ç®—æ•°ç»„ä¸­æ‰€æœ‰æ•°å­—çš„å’Œ",
            "parameters": {
                "type": "object",
                "properties": {
                    "numbers": {
                        "type": "array",
                        "items": {
                            "type": "number"
                        }
                    }
                }
            },
        }],
    )
    return response.choices[0].message
````

````python
def get_completion(messages, model="gpt-4"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,  # æ¨¡å‹è¾“å‡ºçš„éšæœºæ€§ï¼Œ0 è¡¨ç¤ºéšæœºæ€§æœ€å°
        functions=[  # ç”¨ JSON æè¿°å‡½æ•°ã€‚å¯ä»¥å®šä¹‰å¤šä¸ªï¼Œä½†æ˜¯åªæœ‰ä¸€ä¸ªä¼šè¢«è°ƒç”¨ï¼Œä¹Ÿå¯èƒ½éƒ½ä¸ä¼šè¢«è°ƒç”¨
            {
                "name": "sum",
                "description": "è®¡ç®—æ•°ç»„ä¸­æ‰€æœ‰æ•°å­—çš„å’Œ",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "numbers": {
                            "type": "array",
                            "items": {
                                "type": "number",
                                "description": "å¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                            }
                        }
                    }
                },
            },
            {
                "name": "subtract",
                "description": "è®¡ç®— a - b çš„å€¼",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "a": {
                            "type": "number",
                            "description": "è¢«å‡æ•°ï¼Œå¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                        },
                        "b": {
                            "type": "number",
                            "description": "å‡æ•°ï¼Œå¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                        }
                    }
                },
            },
            {
                "name": "multiply",
                "description": "è®¡ç®—æ•°ç»„ä¸­æ‰€æœ‰æ•°å­—çš„ç§¯",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "numbers": {
                            "type": "array",
                            "items": {
                                "type": "number",
                                "description": "å¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                            }
                        }
                    }
                },
            },
            {
                "name": "divide",
                "description": "è®¡ç®— a/b çš„å€¼",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "a": {
                            "type": "number",
                            "description": "è¢«é™¤æ•°ï¼Œå¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                        },
                        "b": {
                            "type": "number",
                            "description": "é™¤æ•°ï¼Œå¿…é¡»æ˜¯æ•°å€¼ç±»å‹"
                        }
                    }
                },
            }
        ],
    )
    return response.choices[0].message
````

````python
prompt = "6 * 3 / (4+2) = ?"
# prompt = "æ¡Œä¸Šæœ‰ 2 ä¸ªè‹¹æœï¼Œå››ä¸ªæ¡ƒå­å’Œ 3 æœ¬ä¹¦ï¼Œæ°´æœæ¯”ä¹¦å¤šå¤šå°‘ï¼Ÿ"
# prompt = """
# è®©æˆ‘ä»¬ä¸€æ­¥æ­¥è®¡ç®—ï¼šå°æ˜åœ¨ä¸€å®¶æ°´æœåº—ä¹°æ°´æœã€‚ä»–ä¹°äº†Xæ–¤è‹¹æœï¼Œæ¯æ–¤10å…ƒï¼›4æ–¤é¦™è•‰ï¼Œæ¯æ–¤5å…ƒï¼›
# å’Œ3æ–¤æ©™å­ï¼Œæ¯æ–¤8å…ƒã€‚ä»–æ‰‹å¤´æœ‰100å…ƒã€‚è¯·é—®å°æ˜ä¹°å®Œè¿™äº›æ°´æœåï¼Œä»–è¿˜å‰©ä¸‹å¤šå°‘é’±ï¼Ÿ
#"""

messages = [
    {"role": "system", "content": "ä½ æ˜¯ä¸€ä¸ªå°å­¦æ•°å­¦è€å¸ˆï¼Œä½ è¦æ•™å­¦ç”Ÿå››åˆ™æ··åˆè¿ç®—"},
    {"role": "user", "content": prompt}
]

response = get_completion(messages)
messages.append(response)  # æŠŠå¤§æ¨¡å‹çš„å›å¤åŠ å…¥åˆ°å¯¹è¯ä¸­ã€‚éå¸¸é‡è¦ï¼
print(response)
while (response.get("function_call")):
    # æ˜¯å¦è¦è°ƒç”¨ sum
    args = json.loads(response["function_call"]["arguments"])
    function_name = response["function_call"]["name"]

    if (function_name == "sum"):
        result = sum(args["numbers"])
    elif (function_name == "subtract"):
        result = args["a"] - args["b"]
    elif (function_name == "multiply"):
        result = 1
        for number in args["numbers"]:
            result *= number
    elif (function_name == "divide"):
        result = args["a"] / args["b"]
    else:
        result = "Unknown function"

    print(result)
    messages.append(
        {"role": "function", "name": function_name, "content": str(result)})
    response = get_completion(messages)
    messages.append(response)  # æŠŠå¤§æ¨¡å‹çš„å›å¤åŠ å…¥åˆ°å¯¹è¯ä¸­
    print(response)

print(response.content)
````

# Function Calling ç¤ºä¾‹ 3ï¼šæŸ¥è¯¢æ•°æ®åº“


éœ€æ±‚ï¼šä»è®¢å•è¡¨ä¸­æŸ¥è¯¢å„ç§ä¿¡æ¯ï¼Œæ¯”å¦‚æŸä¸ªç”¨æˆ·çš„è®¢å•æ•°é‡ã€æŸä¸ªå•†å“çš„é”€é‡ã€æŸä¸ªç”¨æˆ·çš„æ¶ˆè´¹æ€»é¢ç­‰ç­‰ã€‚


````python
def get_sql_completion(messages, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,  # æ¨¡å‹è¾“å‡ºçš„éšæœºæ€§ï¼Œ0 è¡¨ç¤ºéšæœºæ€§æœ€å°
        function_call="auto",
        functions=[{  # æ‘˜è‡ª OpenAI å®˜æ–¹ç¤ºä¾‹ https://github.com/openai/openai-cookbook/blob/main/examples/How_to_call_functions_with_chat_models.ipynb
            "name": "ask_database",
            "description": "Use this function to answer user questions about business. Output should be a fully formed SQL query.",
            "parameters": {
                "type": "object",
                "properties": {
                    "query": {
                        "type": "string",
                        "description": f"""
                            SQL query extracting info to answer the user's question.
                            SQL should be written using this database schema:
                            {database_schema_string}
                            The query should be returned in plain text, not in JSON.
                            """,
                    }
                },
                "required": ["query"],
            },
        }],
    )
    return response.choices[0].message
````

````python
#  æè¿°æ•°æ®åº“è¡¨ç»“æ„
database_schema_string = """
CREATE TABLE orders (
    id INT PRIMARY KEY NOT NULL, -- ä¸»é”®ï¼Œä¸å…è®¸ä¸ºç©º
    customer_id INT NOT NULL, -- å®¢æˆ·IDï¼Œä¸å…è®¸ä¸ºç©º
    product_id INT NOT NULL, -- äº§å“IDï¼Œä¸å…è®¸ä¸ºç©º
    price DECIMAL(10,2) NOT NULL, -- ä»·æ ¼ï¼Œä¸å…è®¸ä¸ºç©º
    status INT NOT NULL, -- è®¢å•çŠ¶æ€ï¼Œæ•´æ•°ç±»å‹ï¼Œä¸å…è®¸ä¸ºç©ºã€‚0ä»£è¡¨å¾…æ”¯ä»˜ï¼Œ1ä»£è¡¨å·²æ”¯ä»˜ï¼Œ2ä»£è¡¨å·²é€€æ¬¾
    create_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP, -- åˆ›å»ºæ—¶é—´ï¼Œé»˜è®¤ä¸ºå½“å‰æ—¶é—´
    pay_time TIMESTAMP -- æ”¯ä»˜æ—¶é—´ï¼Œå¯ä»¥ä¸ºç©º
);
"""

prompt = "ä¸Šä¸ªæœˆçš„é”€å”®é¢æ˜¯å¤šå°‘ï¼Ÿ"
# prompt = "ç»Ÿè®¡æ¯æœˆæ¯ä»¶å•†å“çš„é”€å”®é¢"
# prompt = "è¿™æ˜ŸæœŸå“ªä¸ªç”¨æˆ·æ¶ˆè´¹æœ€é«˜ï¼Ÿæ¶ˆè´¹å¤šå°‘ï¼Ÿ"
messages = [
    {"role": "system", "content": "åŸºäº order è¡¨å›ç­”ç”¨æˆ·é—®é¢˜"},
    {"role": "user", "content": prompt}
]
response = get_sql_completion(messages)
print(response)
````
